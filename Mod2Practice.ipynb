```python
import seaborn as sns
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('companies.csv')
df.head()
```

```python
sns.countplot(x='state', data=df)
plt.show()
```

```python
df.sort_values(by="job_openings", inplace=True)
df.head()
```

```python
plt.plot(df['job_openings'], df['company_name'])
plt.show()
```

```python
job_openings_df = df.groupby('job_openings').agg({'state': 'count'})
job_openings_df
```
