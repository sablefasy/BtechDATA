```python
import seaborn as sns
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('companies.csv')
df.head()
```

```python
sns.countplot(x='state', data=df)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.countplot(x='state', data=df)
```

```python
jobs_industry_df = df.groupby('Industry').agg({'job_openings': 'sum'})
jobs_industry_df
```

```python
sns.barplot(x='job_openings', y='Industry', data=jobs_industry_df)
plt.show()
```

```python
growth_founded_df = df.groupby('founded').agg({'growth_percentage': 'mean'}).reset_index()
growth_founded_df
```

```python
sns.lineplot(x='founded', y='growth_percentage', data=growth_founded_df)
plt.title("growth percentage by year founded")
plt.show()
```

```python
sns.scatterplot(x='employees', y='estimated_revenues', data=df)
plt.title("estimated revenues by number of employees")
plt.show()
```

```python
sns.histplot(x='founded', data=df)
plt.title("Number of Companies Founded per Year")
plt.show()
```
