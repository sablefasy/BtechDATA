```python
import seaborn as sns
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('housing.csv')
df.head()
```

```python
sns.scatterplot(x='Landsize', y='Price', hue='Type', data=df)
plt.title("Housing Prices vs Land Size")
plt.show()
```

```python
avg_price_per_year_df=df.groupby('YearBuilt').agg({'Price':'mean'}).reset_index()
sns.lineplot(x='YearBuilt', y='Price', data=avg_price_per_year_df)
plt.title("Average Housing Price per Year")
plt.show()
```

```python
avg_price_per_year_df=df.groupby('Regionname').agg({'Price':'mean'}).reset_index()
sns.barplot(x='Regionname', y='Price', data=avg_price_per_year_df)
plt.title("Housing Prices by Region")
plt.xticks(rotation=90)
plt.show()
```

```python
sns.histplot(df['Price'], bins=20, kde=True)
plt.title("Distribution of Housing Prices")
plt.show()
```

```python
sns.boxplot(x='YearBuilt', y='Regionname', data=df)
plt.title("Housing Prices by Year and Region")
plt.show()
```
