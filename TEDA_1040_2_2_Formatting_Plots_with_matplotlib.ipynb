# 1.2 Formatting Plots with matplotlib

Matplotlib makes formatting graphs easy. In this notebook, you will learn how to create basic plots and format them in your notebooks.

### About the data

The data used in this notebook is made up and doesn't use a CSV file. This is done on purpose to show how matplotlib works without focusing too much on the data. In a future notebook, we will create plots using data that comes from an external source.

## Using matplotlib
### Importing matplotlib
As explained in the previous reading, the actual *matplotlib* library contains many sets of functions. The main set of functions contained in the library is called `pyplot` and is imported from the matplotlib library under the alias `plt` by convention.

In the code below, we import `pyplot` from the `matplotlib` library under the alias `plt`.

```python
import matplotlib.pyplot as plt
```

### Creating a plot
The `plt` key word can now use functions to reference and create many of the different plots that matplotlib can produce. The most basic of all plots is the **line graph**, which can be created by using the function `plot()`. As we would with any other graph type, we can create a line graph by simply using the `plt` key word and then getting its `plot()` function out in the format `plt.plot()`. Try it out by running the code cell below!

After running the code cell below, notice what happened. Matplotlib created an empty plot (an empty canvas to graph on) and printed it out to the screen. The plot is empty because we didn't pass any data to the `plot()` function. It also looks like matplotlib placed tick marks on the empty canvas by default at intervals of 0.02, with 0.00 being placed in the middle of the y- and x-axes. This is expected since we haven't passed any data to the `plot()` function.

Interestingly, however, something else happened. Right above the empty graph, a set of square brackets was output `[]`. This was output because the `plot()` function actually returns returns a list of lines, and Jupyter Notebook automatically prints out whatever is returned on the last line of each cell. *The `plot()` function is not meant to print out graphs-- it is only meant to draw lines!* The `plot()` function returned an empty list `[]` since it wasn't given any data and no lines were created. Then, even though `plot()` isn't meant to be used for rendering graphs, the empty graph was printed out.

```python
plt.plot()
```

If we pass in data to the `plot()` function, we will see the empty brackets above the empty graph take on a pointer to a `Line2D` object, which is overlaid on the canvas and then printed out.

Notice that the first argument in the `plot()` function is a list with two items, corresponding to the x-axis value for two rows. The second argument is also a list with two items, corresponding to the y-axis value for those same two rows.

```python
plt.plot([1,2], [4, 9])
```

Now we can see a `Line2D` object in the list immediately above the graph. However, **we should choose to not display this list by adding a line `plt.show()` below the `plot()` function**.

```python
plt.plot([1, 2], [4, 9])
plt.show()
```

The `show()` function is essential when using matplotlib and `plt`. Because you may make multiple plots on a single graph, the `show()` function allows you to superimpose multiple plots on top of each other and print them all out on top of each other. For example, I could add another `plot()` function to the code cell above and create another line to be shown on the same plot:

```python
plt.plot([1, 2], [4, 9])
plt.plot([1, 2], [3, 10])
plt.show()
```

**You should always use the `show()` funntion *after* making your plots**. This allows each plot to be created on the same graph and visualized together.

### Adding data to a plot
One of the most important aspect of graphing using matplotlib is understanding the kind of data that your plot requires. In a line graph, for example, both the y-axis and x-axis can contain some kind of numerical data.

Data to be used in matplotlib should first be stored in a list-like structure. For example, let's create the following lists describing iphone sales during the past twelve years:

```python
iphone_sales = [215.05, 190.26, 169.22, 231.22, 211.88, 216.76, 217.72, 220.48, 215.76, 202.31, 230.1, 250.87]
year = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]
```

The first list, `iphone_sales`, contains twelve difference observations describing the sales amounts of iphones. The second list, `year`, contains twelve different years. **Importantly, both lists have the same number of items**.

Using a line graph, we could use these lists to visualize iphone sales over the past twelve years.

The `plot()` function takes two required parameters, `x` and `y`, for graphing a line. The `x` argument will correspond to each value on the x-axis, and `y` will correspond to a value on the x-axis. Both parameters should be lists of data where the first value of the x-axis list corresponds to the first value of the y-axis lists, and so on.

Knowing this, we can plot iphone sales over time by simply calling the `plot()` function and passing in `year` as the x-axis and `iphone_sales` as the y-axis. *Don't forget to `show()`!*

```python
plt.plot(year, iphone_sales)
plt.show()
```

Maybe we also have data from Google about Pixel sales during the past twelve years. With that information, we can create two more lists that describe the years and the sales amounts and superimpose the new line on the graph on iphone sales for easy comparison!

```python
pixel_sales = [208.5, 195.96, 157.37, 252.02, 209.76, 234.10, 219.89, 209.45, 189.86, 188.14, 213.99, 250.87]
pixel_year = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]
```

```python
plt.plot(year, iphone_sales)
plt.plot(pixel_year, pixel_sales)
plt.show()
```

### Formatting a plot
There are many ways that you can format a plot using matplotlib. Below, you will read about some of them.

#### Adding a legend
In the graph above, it's impossible to tell what each color of line represents. To improve this graph, we can create a legend that will label each color to tell us what it represents.

To add a legend, we can simply use the `legend()` function of `plt`. This function doesn't need any parameters. However, if we don't give the plot labels for each line, it won't know what to call them and will throw a warning when we run the cell.

```python
plt.plot(year, iphone_sales)
plt.plot(pixel_year, pixel_sales)
plt.legend()
plt.show()
```

We can fix this by adding labels to each line as we plot it in the `plot()` function. We can add a label by using the `label` parameter of the `plot()` function and setting it equal to whatever we want to the line to be labelled as.

```python
plt.plot(year, iphone_sales, label='iPhone Sales')
plt.plot(pixel_year, pixel_sales, label='Pixel Sales')
plt.legend()
plt.show()
```

##### Change the legend location
Matplotlib is pretty good about knowing where to place the legend in the graph so that it doesn't show on top of other lines. If you want to change it manually, however, you can pass in a parameter `loc` to the `legend()` function and set it equal to a string like "upper right" or "lower center".

```python
plt.plot(year, iphone_sales, label='iPhone Sales')
plt.plot(pixel_year, pixel_sales, label='Pixel Sales')
plt.legend(loc='lower center')
plt.show()
```

#### Adding a title and axis labels
Adding a title and axis labels to the plot is easy. The `plt` key word comes with the functions `title()`, `xlabel()` and `ylabel()` that each take a string to be used as the label. Simply pass in the string to the function and see the labels appear on the plot.

```python
plt.plot(year, iphone_sales)
plt.plot(pixel_year, pixel_sales)

plt.title("iPhone and Pixel Sales For 12 Years")
plt.xlabel("Year")
plt.ylabel("Sales")

plt.show()
```

#### Changing the plot size
Right now, the default plot size is kind of small and a little blurry. We can make the plot bigger by using the `figure()` function, passing in the parameter `figsize`, and setting it equal to a tuple of two integers, representing the height and width.

The code will look something like `plt.figure(figsize=(width, height))`.

```python
plt.figure(figsize=(16, 8))
plt.plot(year, iphone_sales)
plt.plot(pixel_year, pixel_sales)
plt.show()
```

##### Dots per inch
If you notice that the plot is too blurry, you can try to increase the dots per inch of the graphic by passing in the parameter `dpi` set to an integer to the `figure()` function as well.

```python
plt.figure(figsize=(16, 8), dpi=1000)
plt.plot(year, iphone_sales)
plt.plot(pixel_year, pixel_sales)
plt.show()
```

#### Setting the tick marks
In the graph above, it looks like matplotlib defaulted to creating a tick mark only for the even numbered years. We can manually set the tick marks of the plot by using the `xticks()` function. This function accepts two parameters: `ticks`, which is equal to a list of positions, and `labels`, which is set equal to a list of labels for each position.

For example, in the code below we can show each year by passing in our previously created list `year` to the `ticks` argument. We also pass it to the `labels` argument. This is essentially saying that at position 2010, put the label 2010, and so on.

```python
plt.plot(year, iphone_sales)
plt.plot(year, pixel_sales)

plt.xticks(ticks=year, labels=year)

plt.show()
```

We can also pass in strings as labels, as long as the position of each string corresponds with a location specified in the `ticks` parameter. To clarify, labels and ticks must have the same number of items.

```python
plt.plot(year, iphone_sales)
plt.plot(pixel_year, pixel_sales)

plt.xticks(ticks=year, labels=['Beginning', None, None, None, None, None, None, None, None, None, None, 'End'])

plt.show()
```

In another example, we could simply pass in two items to the `ticks` argument to make a tick mark at 2010 and 2021 and then place labels there. These lists both only have two items, since there are only two tick marks.

```python
plt.plot(year, iphone_sales)
plt.plot(pixel_year, pixel_sales)

plt.xticks(ticks=[2010, 2021], labels=['Beginning', 'End'])

plt.show()
```

The same principles apply to using the `yticks()` function to set the y tick marks and labels.

```python
plt.plot(year, iphone_sales, label='iPhone Sales')
plt.plot(pixel_year, pixel_sales, label='Pixel Sales')

plt.yticks(ticks=[160, 250], labels=['Bottom', 'Top'])

plt.show()
```

#### Formatting lines

We can use special parameters when creating our plots to format the visual aspects of the lines in the graph. There are many ways you could potentially format the data points in the graph, but the most common ways to format lines is by changing colors, line styles, markers, and line sizes.

##### Colors
As we create them, we can change the colors of the lines in the graph by using the `color` parameter and setting it equal to color name as a string (ie. `"red"`).

```python
plt.plot(year, iphone_sales, color='red')
plt.plot(pixel_year, pixel_sales, color='green')

plt.show()
```

We can also use hexadecimal colors.

```python
plt.plot(year, iphone_sales, color='#AA544A')
plt.plot(pixel_year, pixel_sales, color='#24ACD3')

plt.show()
```

##### Line style
You can change the line style of the actual line being plotted with the `linestyle` parameter. The values that this parameter can take include `"-"` (solid line), `":"` (dotted line), `"--"` (dashed line), `"-."` (dashdot), and `" "` (no line). Note that these styles are all strings set to the `linestyle` paraemeter.

```python
plt.plot(year, iphone_sales, linestyle="--")
plt.plot(pixel_year, pixel_sales, linestyle=":")

plt.show()
```

##### Markers
Markers refer to the actual points of the data. By default, markers are not shown in line graphs, but can be customized in each line with the parameter `marker`. The parameter can be set to take one of several different string values, including `"o"` (circle), `"*"` (star), `"."` (point), and `"x"` (X).

```python
plt.plot(year, iphone_sales, marker="o")
plt.plot(pixel_year, pixel_sales, marker="x")

plt.show()
```

##### Line width
Sometimes lines are so small that they are hard to see. To fix this, we can use the `linewidth` parameter set to a number to specify how thick each line should be.

```python
plt.plot(year, iphone_sales, linewidth=1.5)
plt.plot(pixel_year, pixel_sales, linewidth=5)

plt.show()
```

### Putting it all together
Now that we've learned about all the different things that we can do to a `matplotlib` plot, let's put it all together and make a graph that uses **all** of the different concepts at once!

Usually, when making graphs with `matplotlib`, our code will look something like what is shown below. This is because there are a lot of tasks that go into making a single graph.

```python
plt.figure(figsize=(10, 8)) # Change plot size

plt.plot(year, iphone_sales, label="iPhone Sales", color="orange", linestyle=":", marker="*", linewidth=5)
plt.plot(pixel_year, pixel_sales, label="Pixel Sales", color="#46AA23", linestyle="--", marker="o", linewidth=5)

plt.legend() # Add a legend

plt.title("iPhone and Pixel Sales For 12 Years")
plt.xlabel("Year")
plt.ylabel("Sales")

plt.xticks(ticks=year, labels=['Beginning', None, None, None, None, None, None, None, None, None, None, 'End'])
plt.yticks(ticks=[180, 190, 230, 250])

plt.show()
```
