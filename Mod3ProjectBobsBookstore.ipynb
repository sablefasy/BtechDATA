```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get('https://btech-data-analytics.github.io/bridgerland-technical-college/bookstore.html').text
```

```python
soup =  BeautifulSoup(response, 'html.parser')
```

```python
soup.find('table').find_all('tr')
```

```python
# prompt: create a dataframe with author's name and number of books each author has as well as a column showing topic, price and number of pages

import pandas as pd
# Find the table and extract rows
table = soup.find('table')
rows = table.find_all('tr')

# Extract headers
headers = [header.text.strip() for header in rows[0].find_all('th')]

data = []
for row in rows[1:]:
    row_data = [cell.text.strip() for cell in row.find_all('td')]
    data.append(row_data)

df = pd.DataFrame(data, columns=headers)

df = df.rename(columns={
    'Author': 'author',
    'Title': 'book',
    'Topic': 'topic',
    'Price': 'price',
    'Pages': 'number of pages'
})

df['price'] = pd.to_numeric(df['price'].replace({'\$': ''}, regex=True), errors='coerce')
df['number of pages'] = pd.to_numeric(df['number of pages'], errors='coerce')

author_book_counts = df.groupby('author').size().reset_index(name='number of books')

df_result = pd.merge(df, author_book_counts, on='author')

df_result = df_result[['author', 'number of books', 'book', 'topic', 'price', 'number of pages']]

df_result = df_result.drop_duplicates(subset=['author', 'book'])

df_result
```

```python
average_pages_per_topic = df.groupby('topic')['number of pages'].mean().reset_index()

average_pages_per_topic = average_pages_per_topic.rename(columns={'number of pages': 'average number of pages'})

print("Average number of pages per topic:")
average_pages_per_topic
```
