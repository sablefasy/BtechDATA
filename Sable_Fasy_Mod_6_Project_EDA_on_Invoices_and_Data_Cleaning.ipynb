```python
df.drop(['LastEditedWhen'], axis=1, inplace=True)
df.isnull().sum()
```

```python
df[df['StockItemID'].isnull()&df['Description'].isnull()]
```

```python
df = df[~(df['StockItemID'].isnull()&df['Description'].isnull())]
```

```python
df.shape
```

```python
df.describe()
```

```python
df['TaxRate'] = df['TaxRate'].abs()
df.describe()
```

```python
df['TaxAmount'].fillna( (df['UnitPrice'] * df['Quantity'] * df['TaxRate']) / 100, inplace=True)
df.head()
df.describe()
```

```python
df.loc[df['ExtendedPrice'] == -999, 'ExtendedPrice'] = (
    (df['UnitPrice'] * df['Quantity']) + df['TaxAmount']
)
```

```python
df['ExtendedPrice'].fillna( (df['UnitPrice'] * df['Quantity']) + df['TaxAmount'], inplace=True)
df['ExtendedPrice'].mean()
```

```python
df.describe()
```

```python
df['ZScoreQuantity'] = (df['Quantity'] - df['Quantity'].mean()) / df['Quantity'].std()
df.head()
```

```python
df['ZScoreEP'] = (df['ExtendedPrice'] - df['ExtendedPrice'].mean()) / df['ExtendedPrice'].std()
df.head()
```

```python
df[df['ZScoreQuantity'] >= 3]
```

```python
df[df['ZScoreEP'] >= 3]
```

```python
df.drop_duplicates(subset=['InvoiceLineID'], inplace=True)
df.isna().sum()
```

```python
df.info()
```

```python
quantity_by_description = df.groupby('Description')['Quantity'].sum()
highest_quantity_item = quantity_by_description.idxmax()
highest_quantity = quantity_by_description.max()

print(f"The item with the highest total quantity sold is '{highest_quantity_item}' with a total quantity of {highest_quantity}.")
```

```python
profit_by_description = df.groupby('Description')['LineProfit'].sum()
highest_profit_item = profit_by_description.idxmax()
highest_profit = profit_by_description.max()

print(f"The item with the highest total profit is '{highest_profit_item}' with a total profit of {highest_profit}.")
```

```python
average_quantity_per_order = df.groupby('Description')['Quantity'].mean()
highest_average_quantity_item = average_quantity_per_order.idxmax()
highest_average_quantity = average_quantity_per_order.max()

print(f"The item purchased in the greatest quantities per order, on average, is '{highest_average_quantity_item}' with an average quantity of {highest_average_quantity}.")
```
